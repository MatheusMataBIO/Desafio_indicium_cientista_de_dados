```python
# Gráfico de barras
plt.figure(figsize=(10,6))
top_directors = df_indicium['Director'].value_counts().head(10)
sns.barplot(x=top_directors.values, y=top_directors.index, palette="magma")
plt.title("Top 10 Diretores com Mais Filmes no Dataset", fontsize=14)
plt.xlabel("Quantidade de Filmes")
plt.ylabel("Diretor")
plt.show()
```

```python
# Ordenar classificações pela mediana da receita
order_median = df_indicium.groupby("Certificate")["Gross"].median().sort_values(ascending=False).index

# Gráfico boxplot
plt.figure(figsize=(10,6))
sns.boxplot(data=df_indicium, x="Certificate", y="Gross", order=order_median, palette="viridis")

plt.xticks(rotation=60)
plt.title("Distribuição da Receita Bruta por Classificação Indicativa", fontsize=14)
plt.ylabel("Receita Bruta")
plt.xlabel("Classificação Indicativa")
plt.tight_layout()
plt.show()


```

```python
# Selecionando apenas colunas numéricas
num_cols = df_indicium.select_dtypes(include=['float64', 'int64'])

# Matriz de correlação de Pearson
corr = num_cols.corr(method="pearson")

plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

plt.title("Matriz de Correlação (Pearson)", fontsize=14)
plt.show()

```

```python
avg_ratings = df_indicium.groupby("Genre")["IMDB_Rating"].mean().sort_values()

# Pegar top 5 melhores e menores gêneros
best_worst = pd.concat([avg_ratings.head(5), avg_ratings.tail(5)])

# Gráfico de barras
plt.figure(figsize=(8,5))
sns.barplot(x=best_worst.values, y=best_worst.index, palette="coolwarm")

plt.title("Top 5 melhores e piores gêneros por média de nota", fontsize=14)
plt.xlabel("Média IMDB Rating")
plt.ylabel("Gênero")
plt.show()



```
