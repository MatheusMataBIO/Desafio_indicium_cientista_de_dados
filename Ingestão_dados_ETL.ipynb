```python
import pandas as pd

# Ler o arquivo CSV
df_indicium = pd.read_csv("desafio_indicium_imdb.csv")
df_indicium.display()
```

```python
# Remover a coluna "Unnamed: 0" pois não tem muita utilidade para o projeto
df_indicium = df_indicium.drop(columns=["Unnamed: 0"])
```

```python
# Converter 'Runtime' para numérico
df_indicium["Runtime"] = df_indicium["Runtime"].str.replace(" min", "", regex=False).astype(float)
```

```python
import numpy as np

# Converter 'Gross' para numérico
df_indicium["Gross"] = (
    df_indicium["Gross"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .replace("nan", np.nan)
    .astype(float)
)
```

```python
df_indicium.info()
```

```python
# Converter 'Released_Year' para numérico
df_indicium['Released_Year'] = pd.to_numeric(df_indicium['Released_Year'], errors='coerce')

# Preencher valores ausentes em 'Certificate' com 'Unrated'
df_indicium['Certificate'] = df_indicium['Certificate'].fillna('Unrated')

# Preencher 'Meta_score' com a média
df_indicium['Meta_score'] = df_indicium['Meta_score'].fillna(df_indicium['Meta_score'].mean())

# Preencher 'Gross' com a mediana
df_indicium['Gross'] = df_indicium['Gross'].fillna(df_indicium['Gross'].median())

# Converter 'Runtime' para inteiro
df_indicium['Runtime'] = df_indicium['Runtime'].astype(int)

```

```python
# Salvando o dataset limpo em CSV
df_indicium.to_csv("filmes_tratado.csv", index=False)
```
